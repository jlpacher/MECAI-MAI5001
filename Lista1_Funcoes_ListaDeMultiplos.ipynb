```python
# Escreva um programa em Python que retorna os números divisíveis por x
# e múltiplos de y no intervalo [Mín, Máx], onde Min, Máx, x e y são fornecidos como entrada.
# Utilize a função eh_divisivel(a,b) para retornar verdadeiro se um número a é divisível por b.
# A entrada apresenta os valores Min, Máx, x e y
# Dica: Use ','.join() para formatar a saída
# inputs: 100 200
# 5,7
# outputs: 105,140,175
import time

minimo,maximo =  map(int,input().split(','))
denominador1,denominador2 = map(int,input().split(','))

def eh_divisivel(a,b):
  if a%b==0:
    return True
  else:
    return False

def calcula_multiplos(minimo,maximo,denominador1,denominador2):
  result=[]
  for number in range(minimo,maximo+1):
    if eh_divisivel(number,denominador1) and eh_divisivel(number,denominador2):
      result.append(str(number))
  return ",".join(result)

print(calcula_multiplos(minimo,maximo,denominador1,denominador2))
time.sleep(0.8)
```

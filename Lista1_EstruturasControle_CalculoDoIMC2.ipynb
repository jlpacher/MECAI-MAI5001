```python
# Modifique o código para calculo do Índice de Massa Corporal (IMC) para também classificar o IMC obtido.
# Para isso, considere a seguinte classificação do IMC segundo a OMS:
# Classificação do IMC de acordo com a OMS:
# IMC <18,5kg/m2 - baixo peso.
# IMC >18,5 até 24,9kg/m2 - eutrofia (peso adequado)
# IMC ≥25 até 29,9kg/m2 - sobrepeso.
# Lembrando que: IMC = m/(altura)^2 (massa dividida pela altura ao quadrado)
# IMC = m/(altura)^2 massa dividida pela altura ao quadrado
# inputs  1.75,80.3    # 1.63,76.5    # 1.91,93.3    # 2.03,85.4
# outputs 26.22 28.79 25.57 20.72
import time
from math import floor

alturas = []
massas = []

while True:
  try:
    altura,massa = input().split(",")
    alturas.append(float(altura))
    massas.append(float(massa))
  except:
    break

def floor_2_decimals(number):
  # prepare number to apply floor method
  number = number*100
  # apply floor method
  number = floor(number)
  # rebuild 2 decimals
  number = number/100
  return number

imc = []
for i in range(len(alturas)):
  altura,massa = alturas[i],massas[i]
  calculo = massa/(altura**2)
  calculo_ajustado = floor_2_decimals(calculo)
  imc.append(calculo_ajustado)

imc = [ '%.2f' %elem for elem in imc ]

for elem in imc:
  print(f'{elem} ', end='')
  if round(float(elem),1)<18.5:
    print('Baixo peso')
  elif round(float(elem),1)>=18.5 and round(float(elem),1)<=24.9:
    print('Peso adequado')
  else:
    print('Sobrepeso')

time.sleep(0.8)
```

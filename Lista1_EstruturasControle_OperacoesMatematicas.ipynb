```python
# Desenvolva um programa em python que realize a operação fornecida na entrada de dados.
# Os códigos das operações são: + (soma), - (subtração), * (multiplicação), / (divisão) e ^ (potência). O código deve calcular o resultado da operação.
# inputs  2*3
# outputs 6

line = input() #.split(",")

if '+' in line:
  a=int(line.split('+')[0])
  b=int(line.split('+')[1])
  result=a+b

elif '*' in line:
  a=int(line.split('*')[0])
  b=int(line.split('*')[1])
  result=a*b

elif '/' in line:
  a=int(line.split('/')[0])
  b=int(line.split('/')[1])
  if b==0:
    result='Erro: divisão por zero'
  else:
    result=round(a/b,3)

elif '^' in line:
  a=int(line.split('^')[0])
  b=int(line.split('^')[1])
  if a==0 and b==0:
    result='Erro: base e expoente nulos'
  else:
    result=a**b

elif '-' in line:
  if '-'not in line[0]:
      a=int(line.split('-')[0])
      b=int(line.split('-')[1])
      result=a-b
  else:
      a=int(line.split('-')[1])
      b=int(line.split('-')[2])
      result=a-b

else:
  result='Erro: nenhuma operação possível'

print(result)
```

```python
# Desenvolva um código, usando funções, que retorne o resultado das séries abaixo.
# Uma série pode ser implementada iterativamente,
# mas a outra deve ser implementada recursivamente como descrito abaixo:
# Iterativamente: s1(x) = x - x^3/3! + x^5/5! - x^7/7! +...
# Recursivamente: s2(x) = 1 - x^2/2! + x^4/4! - x^6/6! +...
# As entradas fornecidas são os valores x e n (tamanho da série)
# inputs: 1 5
# outputs: 0.841   0.540
from math import factorial as fact

x,n = input().split()
x = float(x)
n = int(n)

def s1(x,n):
  '''
  Calcula uma série iterativamente
  entradas: x,n (variável e tamanho da série)
  saída: resultado da série
  '''
  # define o sinal do multiplicador de cada termo
  sign=1
  # guarda o resultado
  result = 0
  # itera sobre os termos da série
  for i in range(1,n*2,2):
      result += x**i/fact(i)*sign
      sign *= (-1)
  return result

def s2(x,n):
  '''
  Calcula uma série recursivamente
  entradas: x,n (variável e tamanho da série)
  saída: resultado da série
  '''
  # define a parada
  if n==0:
    return 0
  # calcula a série recursivamente
  else:
    # define o sinal do multiplicador de cada termo
    if n%2==0:
      sign = -1
    else:
      sign = 1
    return (x**(2*n-2)/fact(2*n-2)) * sign + s2(x,n-1)

print('%.3f' %s1(x,n))
print('%.3f' %s2(x,n))
```

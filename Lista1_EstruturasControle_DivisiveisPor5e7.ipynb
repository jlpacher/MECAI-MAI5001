```python
# Escreva um programa em Python que retorna os números divisíveis por 7 e
# múltiplos de 5 no intervalo [Mín, Máx], onde Min e Máx são fornecidos como entrada.
# Dica: Use ','.join() para formatar a saída
# inputs: 100 200
# outputs: 105,140,175
import time

minimo,maximo =  map(int,input().split(','))

result=[]
for number in range(minimo,maximo+1):
  if number%5==0:
    if number%7==0:
      result.append(str(number))

print(",".join(result))

time.sleep(0.8)
```

```python
# Crie uma classe para números complexos, ou seja, teremos um objeto x com parte real e imaginária: x.real e x.imag
# Implemente as operações a seguir como métodos da classe complexo a ser criada: +, -, /, *, abs, ==.
# Considere a seguinte representação para complexos x=(x.real) + (x.imag)i nas operações detalhadas abaixo
# xy= (x.realy.real - x.imagy.imag) + ( x.imagy.real + x.real*y.imag)i
# x/y= ( (x.realy.real+x.imagy.imag)/r ) + ( (x.imagy.real-x.realy.imag)/r )i, onde r= y.real2 + y.imag2)
# abs(x) = sqrt(x.real2 + x.imag2)
# x==y: Se (x.real==y.real) E (x.imag==y.imag), verdeiro; caso contrário, falso.
# inputs: == abs +
# outputs: 1,2 \n 3,4

class Complexo(object):
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    def __add__(self, outro):
        somaReal = self.real + outro.real
        somaImag = self.imag + outro.imag
        return Complexo(somaReal, somaImag)

    def __sub__(self, outro):
        subtReal = self.real - outro.real
        subtImag = self.imag - outro.imag
        return Complexo(subtReal, subtImag)

    def __mul__(self, outro):
        multReal = (self.real * outro.real) - (self.imag * outro.imag)
        multImag = (self.imag * outro.real) + (self.real * outro.imag)
        return Complexo(multReal, multImag)

    def __truediv__(self, outro):
        r = outro.real ** 2 + outro.imag ** 2
        divReal = ((self.real * outro.real) + (self.imag * outro.imag)) / r
        divImag = ((self.imag * outro.real) - (self.real * outro.imag)) / r
        return Complexo(divReal, divImag)

    def __abs__(self):
        absReal = (self.real ** 2 + self.imag ** 2) ** 0.5
        return absReal

    def __eq__(self, outro):
        if self.real == outro.real and self.imag == outro.imag:
            return True
        return False

    def __str__(self):
        if self.imag>=0:
          return str('%.2f' % self.real) + '+' + str('%.2f' % self.imag) + 'i'
        else:
          return str('%.2f' % self.real) + ' ' + str('%.2f' % self.imag) + 'i'


operador = input().split()
real1, imag1 = map(float, input().split(','))
real2, imag2 = map(float, input().split(','))

# Exemplo de uso:
num1 = Complexo(real1, imag1)
num2 = Complexo(real2, imag2)

for op in operador:
    if op == '+':
        resultado = num1 + num2
        print(f'{op}:{resultado}')
    elif op == '-':
        resultado = num1 - num2
        print(f'{op}:{resultado}')
    elif op == '/':
        resultado = num1 / num2
        print(f'{op}:{resultado}')
    elif op == '*':
        resultado = num1 * num2
        print(f'{op}:{resultado}')
    elif op == 'abs':
        resultado1 = '%.2f' % abs(num1)
        resultado2 = '%.2f' % abs(num2)
        print(f'{op}:{resultado1}')
        print(f'{op}:{resultado2}')
    elif op == '==':
        resultado = num1 == num2
        print(f'{op}:{resultado}')
    else:
        print("Operador inválido")

```

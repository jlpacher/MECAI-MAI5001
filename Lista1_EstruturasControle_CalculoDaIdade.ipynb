```python
# Desenvolva um algoritmo que leia a data de nascimento (dd/mm/aaaa) de uma pessoa e a data atual, calcule e escreva a idade da pessoa.
# Considere a data atual como 04/04/2024 para que os resultados coincidam com a saída fornecida pelo run.codes
# inputs  23/02/1988
# outputs 36 anos, 1 mes e 12 dias
import time
import datetime
from dateutil import relativedelta

born=input()
reference_day='04/04/2024'

def days_between(d1, d2):
  start_date = datetime.datetime.strptime(d1, "%d/%m/%Y")
  end_date = datetime.datetime.strptime(d2, "%d/%m/%Y")
  delta = relativedelta.relativedelta(end_date, start_date)
  return delta.years, delta.months, delta.days

years=days_between(born,reference_day)[0]
months=days_between(born,reference_day)[1]
days=days_between(born,reference_day)[2]+1

if years<=1:
  print(f'{years} ano, ', end='')
else:
  print(f'{years} anos, ', end='')

if months<=1:
  print(f'{months} mes e ', end='')
else:
  print(f'{months} meses e ', end='')

if days<=1:
  print(f'{days} dia')
else:
  print(f'{days} dias')
```

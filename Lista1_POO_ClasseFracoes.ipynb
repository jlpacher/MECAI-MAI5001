```python
# Utilizando a classe fração apresentada em aula, implementar as operações de
# multiplicação (), divisão (/), potência (*), soma (+), subtração (-) e igualdade (==)
# sobrescrevendo os respectivos operadores em Python.
# Veja a figura abaixo que ilustra a série: n=3 => soma=1+1/2+1/3
# inputs: +\n1/2\n5/4     **\n3/7\n5          ==\n1/3\n7/9
# outputs: 7/4            243/16807           False

class Fracao(object):

   def __init__(self, num,den):
       self.num=num
       self.den=den
       self.simplificar()

   def __add__(self,outro):
       somaDen=self.den*outro.den
       somaNum=outro.den*self.num + self.den*outro.num
       if(somaNum%somaDen==0):
           somaNum=somaNum//somaDen
           somaDen=1
       return Fracao(somaNum,somaDen)

   def __sub__(self,outro):
       subtDen=self.den*outro.den
       subtNum=outro.den*self.num - self.den*outro.num
       if(subtNum%subtDen==0):
           subtNum=subtNum//subtDen
           subtDen=1
       return Fracao(subtNum,subtDen)

   def __mul__(self,outro):
       multDen=self.den*outro.den
       multNum=self.num*outro.num
       if(multNum%multDen==0):
           multNum=multNum//multDen
           multDen=1
       return Fracao(multNum,multDen)

   def __truediv__(self,outro):
       divDen=self.den*outro.num
       divNum=self.num*outro.den
       if(divNum%divDen==0):
           divNum=divNum//divDen
           divDen=1
       return Fracao(divNum,divDen)

   def __pow__(self,outro):
       potDen=self.den**(outro.num/outro.den)
       potNum=self.num**(outro.num/outro.den)
       if(potNum%potDen==0):
           potNum=potNum//potDen
           potDen=1
       return Fracao(potNum,potDen)

   def __eq__(self,outro):
       if(self.num/self.den==outro.num/outro.den):
           return True
       return False

   def mdc(self):
       num=int(self.num)
       den=int(self.den)
       resto = num % den
       while(resto):
        num=den
        den=resto
        resto=num%den
       return den

   def simplificar(self):
       divisor= self.mdc()
       self.num=int(self.num/divisor)
       self.den=int(self.den/divisor)

   def __str__(self):
       return str(self.num)+'/'+str(self.den)


num1,den1 = map(int,input().split('/'))
num2,den2 = map(int,input().split('/'))
operador = input()

# Exemplo de uso:
frac1 = Fracao(num1, den1)
frac2 = Fracao(num2, den2)

if operador == '+':
    resultado = frac1 + frac2
elif operador == '-':
    resultado = frac1 - frac2
elif operador == '/':
    resultado = frac1 / frac2
elif operador == '*':
    resultado = frac1 * frac2
elif operador == '**':
    resultado = frac1 ** frac2
elif operador == '==':
    resultado = frac1 == frac2
else:
    print("Operador inválido")

print(resultado)
```

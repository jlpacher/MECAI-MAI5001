```
##Exemplo: Extraindo valores
def quocienteEresto(x,y):
   q=x//y
   r=x%y
   return q,r

print(quocienteEresto(13,7))
x,y=quocienteEresto(13,7)
print(f'{x=},{y=}')
```

```
#Exemplo min_max_numPalavras
def min_max_numPalavras(x):
   numeros=()
   palavras=()
   for t in x:
       numeros=numeros+(t[0],)
       if t[1] not in palavras:
           palavras=palavras+(t[1],)
       print(f'{t=} {numeros=} {palavras=}')
   menor=min(numeros)
   maior=max(numeros)
   numPalavras=len(palavras)
   return menor,maior,numPalavras
t1=((1,"a"),(10,"b"),(5,"b"),(3,"c"),(9,"c"))
t2=min_max_numPalavras(t1)
print(t2)
```

```
#Exemplo: Manipulando listas
a_l=[]
print(a_l)
a_l=[1,2,3,[4,5]]
print(a_l)
a_l[0]=-1
print(a_l)
a_l[3]=6
print(a_l)
a_l[1]='João'
print(a_l)
a_l[2]=3.5
print(a_l)
```

```
#Exemplo: Manipulando listas
a_l=[10, 20, 30, 40, 50]
soma=0
for i in range(len(a_l)):
   print(f' a_l[{i}]={a_l[i]}', end='')
   soma = soma + a_l[i]
print(f' Média={soma/len(a_l)}')

###Python Style
soma=0
for i in a_l:
   print('',i,'',end='')
   soma = soma + i
print(soma/len(a_l))
```

```
#Exemplo: Concatenando
inicio=1
fim=10
a_l=[]
for i in range(inicio,fim):
   a_l=a_l+[i**2]
   print(a_l)
print(a_l)
```

```
#Exemplo: Listas - atenção aos índices
inicio=1
fim=5
a_l=[]
for i in range(inicio,fim):
   print(i)
   a_l=a_l+[i**2]
print(a_l)
print(a_l[4])
a_l[4]=81
print(a_l[4])
```

```
#Exemplo: Adicionando valores à lista
a_l=[1,2,3]
print(a_l)
a_l.append(4)
print(a_l)
a_l.append(5)
print(a_l)
a_l.extend([6,7,8])
print(a_l)
a_l.append([9,10])
print(a_l)
a_l.extend([9,10])
print(a_l)
```

```
#Exemplo: Removendo da listas
a_l=[10,20,30,40,20,50]
print(a_l)
del a_l[2]
print(a_l)
x=a_l.pop(2)
print(x,a_l)
a_l.remove(20)
print(a_l)
#a_l.remove(30)
```

```
#Exemplo: Conversão String e Lista
a_str="Joao"
print(a_str)
x=list(a_str)
print(x)
a_str+=' e Maria'
print(a_str)
x=a_str.split()
print(x)
x=a_str.split('e')
print(x)
a_l=['@','#','!','x']
print(a_l)
a_str=''.join(a_l)
print(a_str)
a_str='_'.join(a_l)
print(a_str)
a_str=':'.join(a_l)
print(a_str)
```

```
#Exemplo: Outras funções e métodos
a_l=[22,15,13,20]
print(a_l)
x=sorted(a_l)
print(x)
print(x[0],a_l[0])
a_l.sort()
print(a_l)
print(x[0],a_l[0])
a_l.reverse()
print(a_l)
a_l=[22,15,13,20]
print(a_l)
a_l.reverse()
print(a_l)
```

```
#Exemplo:Alias
x=[1,2,3,4]
y=x
print(f'{x=}\n{y=}\n')
y.append(5)
print(f'{x}\n{y}')
a=2
b=a
print(a,b)
b=2**2
print(a,b)
```

```
#Exemplo: Alias
x=[1,2,3,[4,5],"Joao"]
y=[x]
print(f'{x=}\n{y=}')
y.extend(x)
print(f'{x=}\n{y=}')
x.append(4)
print(f'{x=}\n{y=}')
x[3].append(6)
print(f'{x=}\n{y=}')
```

```
#Exemplo: Alias
x=[1,2,3,[4,5],"Joao"]
y=[x[:]]
print(f'{x=}\n{y=}')
y.extend(x)
print(f'{x=}\n{y=}')
x.append(4)
print(f'{x=}\n{y=}')
x[3].append(6)
print(f'{x=}\n{y=}')
```

```
#Exemplo: Alias
nomes=['João','Maria','Ana']
idades=[30,25,35]          # nomes                idades
dados_1=[nomes,idades] #[['João','Maria','Ana'],[30,25,35]]
dados_2=[['João','Maria','Ana'],[30,25,35]]
print(f'{dados_1==dados_2}')
print(f'{id(dados_1)==id(dados_2)}')
print(f'{id(dados_1[0])==id(nomes)}')
print(f'{id(dados_1[1])==id(idades)}')
print(f'{id(dados_2[0])==id(dados_1[0])}')

```

```
#Exemplo: Clone
x=[1,2,3]
y=[x[:]]
print(f'{x=}\n{y=}')
x.append(4)
y.append(5)
print(f'{x=}\n{y=}')
z=x[:]
w=list(x)
import copy
u=copy.deepcopy(x)
print(f'{x=},{z=},{w=},{u=}')
print(f'{id(x)},{id(z)},{id(w)},{id(u)}')
x.append(5)
print(f'{x=},{z=},{w=},{u=}')
```

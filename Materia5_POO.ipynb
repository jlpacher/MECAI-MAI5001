```
# #Exemplo 1 - Ponto
class Ponto(object):

    def __init__(self,x,y):
        self.x=x
        self.y=y

    def distancia(self,outro):
        deltaX=(self.x-outro.x)
        deltaY=(self.y-outro.y)
        return (deltaX**2+deltaY**2)**0.5
    def __str__(self):
        return '('+str(self.x)+','+str(self.y)+')'

p=Ponto(3,4)
print(p)
q=Ponto(4,8)
print(p,q)
print(p.distancia(q))
print(q.distancia(p))
```

#Acesso aos atributos


* Não há mecanismos em Python que impeçam de fato o acesso a qualquer variável ou método de uma classe.
* Todos os atributos de uma classe em Python são públicos por definição.
* Atributos públicos (public) podem ser acessados fora da classe.
* Python permite simular acessos do tipo protegido ou privado através da adição de prefixos aos nomes de variáveis e métodos de uma classe.






```
#Exemplo 1- Atributos públicos
class Fracao(object):

   def __init__(self, num,den):
       assert type(num) == int and type(den) == int,'Erro: não é inteiro'
       self.num=num
       self.den=den
       self.simplificar()

   def __add__(self,outro):
       somaDen=self.den*outro.den
       somaNum=outro.den*self.num + self.den*outro.num
       if(somaNum%somaDen==0):
           somaNum=somaNum/somaDen
           somaDen=1
       return Fracao(somaNum,somaDen)

   def __eq__(self,outro):
       if(self.num/self.den==outro.num/outro.den):
           return True
       return False

   def mdc(self):
       num=int(self.num)
       den=int(self.den)
       resto = num % den
       while(resto):
           num=den
           den=resto
           resto=num%den
       return den

   def simplificar(self):
       divisor= self.mdc()
       self.num=int(self.num/divisor)
       self.den=int(self.den/divisor)


   def __str__(self):
       return str(self.num)+'/'+str(self.den)

f1=Fracao(3,4)
print(f1)
f2=Fracao(4,8)
print(f2)
# # print(f1,f2)
# # f3=f1+f2
# # print(f3)
# # print(f1==f2)
# # f3.simplificar()
# # print(f3)

f1.num=8
f1.simplificar()
print(f1)
f2.den=7
print(f2)
f3=f1.__add__(f2) #f3 = f1 + f2
print(f3.__str__())
print(f1.__eq__(f2))
f3.simplificar()
print(f3.__str__())


```

# Classes Abstratas

* As classes abstratas possuem um ou mais métodos do tipo abstrato.
* Um método abstrato é declarado, mas não é implementado.
`@abstractmethod`
* O decorador `@abstractmethod `torna o método abstrato.
* Python por default não fornece classes abstratas.
* Deve-se importar a Abstract Base Classes (ABCs):
`from abc import ABC`


```
#Exemplo 4 - Classes Abstratas

from abc import ABC, abstractmethod

class Calcula(ABC):

  def __init__(self,valor):
      self.valor=valor
  def getValor(self):
      return self.valor
  @abstractmethod
  def calculaValor(self):
      pass

class CalculaDobrando(Calcula):
   def calculaValor(self,valor):
       self.valor= 2*valor

class CalculaSomando(Calcula):

   def calculaValor(self):
       self.valor += 100

class CalculaDividindo(Calcula):
   def calculaValor(self):
     self.valor /=2

   def __str__(self):
      return 'Olá, meu valor é:'+str(self.valor)

# x=Calcula(10)
# print(x.getValor())

x=CalculaDobrando(10)
print(x.getValor())
x.calculaValor(20)
print(x.getValor())

y=CalculaSomando(5)
print(y.getValor())
y.calculaValor()
print(y.getValor())

z=CalculaDividindo(10)
print(z.getValor())
z.calculaValor()
print(z.getValor())



```

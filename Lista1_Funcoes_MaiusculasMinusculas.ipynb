```python
# Escreva uma função em Python que recebe uma string e retorna o númeo de letras maiúsculas e minúsculas.
# inputs: João e Maria
# outputs: Maiúscula: 2   \n Minúscula: 9
import time

texto = input()

def conta_maiusculas_e_minusculas(texto):
  texto = ''.join(e for e in texto if e.isalpha())
  maiusculas,minusculas = 0,0

  for elemento in texto:
    if elemento.isupper():
      maiusculas += 1
    elif elemento.islower():
      minusculas += 1
    else:
      continue

  print(f'Maiúsculas: {maiusculas}\nMinúsculas: {minusculas}')

conta_maiusculas_e_minusculas(texto)
time.sleep(0.8)
```

```python
# Desenvolva um programa em pýthon que leia a altura (metros) e a massa (kg) de uma certa quantidade de pessoa,
# calcule e escreva o Índice de Massa Corporal (IMC) de cada uma delas.
# IMC = m/(altura)^2 massa dividida pela altura ao quadrado
# inputs  1.75,80.3    # 1.63,76.5    # 1.91,93.3    # 2.03,85.4
# outputs 26.22 28.79 25.57 20.72
import time

alturas = []
massas = []

while True:
  try:
    altura,massa = input().split(",")
    alturas.append(float(altura))
    massas.append(float(massa))
  except:
    break

imc = []
for i in range(len(alturas)):
  altura,massa = alturas[i],massas[i]
  calculo = massa/(altura**2)
  imc.append(calculo)

imc = [ '%.2f' %elem for elem in imc ]

print(*imc)
time.sleep(0.8)
```

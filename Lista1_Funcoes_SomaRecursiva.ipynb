```python
# Desenvolva um programa em Python que some valores numa lista usando recursividade.
# inputs: 1 2 3 4 5
# outputs: 15
import time

entrada = map(int,input().split())
numeros = (list(entrada))

def soma_recursiva(lista_de_entrada):
  if not lista_de_entrada:
    return 0
    # Caso recursivo: some o primeiro elemento da lista com a soma dos restantes
  else:
    return lista_de_entrada[0]+soma_recursiva(lista_de_entrada[1:])

# # soma=0
# # for number in numeros:
# #   soma += number

print(soma_recursiva(numeros))
time.sleep(0.8)
```

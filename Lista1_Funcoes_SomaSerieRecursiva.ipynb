```python
# Crie uma função recursiva que retorne a soma de uma séria de tamanho n.
# Veja a figura abaixo que ilustra a série: n=3 => soma=1+1/2+1/3
# inputs: 7
# outputs: 2.592
import time

n = int(input())

def soma_recursiva(n):
  if n==0:
    return 0
  else:
    return 1/n+soma_recursiva(n-1)

print(round(soma_recursiva(n),3))
time.sleep(0.8)
```
